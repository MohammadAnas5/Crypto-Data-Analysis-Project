```python
for i in range(1,538): 
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=d6d6db9828ca83b2e849b84cf888f842&language=en-US&page={i}")
    tempdf = pd.DataFrame(response.json()['results'])[['id','original_title','overview','popularity','release_date','vote_average','vote_count']]  
    df = df.append(tempdf,ignore_index=True)  
```
